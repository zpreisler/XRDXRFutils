### New $\gamma$

To mitigate the negative $\gamma$ instability 

```python
from numpy import linspace,sqrt,sign
from matplotlib.pyplot import plot,subplots,savefig
```

```python
%matplotlib inline

def fnewgamma(x):
    return 0.5 * sqrt(x**2 + 1) + 0.5 * x

def dnewgamma(x):
    return 0.5 + 0.5 *x / sqrt(x**2 + 1)

def huber(x,delta):
    return delta * (sqrt(1 + (x/delta)**2) - 1)

def dhuber(x,delta):
    return x/(delta*sqrt(x**2/delta**2 + 1))

x = linspace(-30,30,10000)

fig,ax = subplots(1,2,figsize=(9,2.5))
fig.subplots_adjust(wspace=0.25,bottom=0.15)

ax[0].plot(x,x,'k--',label=r'$x$',lw=1)

ax[0].set_xlim(-5,5)
ax[0].set_ylim(-0,5)

ax[0].plot(x,fnewgamma(x) ,'-',lw=1.33,label=r'$\frac{1}{2} \sqrt{\gamma^2 + 1} + \frac{1}{2} \gamma$')
ax[0].legend(frameon=False)
ax[0].set_xlabel(r'$\gamma$')
ax[0].set_ylabel(r'$\gamma_\mathrm{new}$')

ax[1].plot(x,sign(x),'k--',lw=1,label=r'$sgn(x)$')
ax[1].plot(x,x*0.5+0.5,'k-',lw=1,label=r'$\frac{1}{2} x + \frac{1}{2}$')

ax[1].set_xlim(-5,5)
ax[1].set_ylim(-0.1,1.1)

ax[1].plot(x,dnewgamma(x) ,'-',lw=1.33,label=r'$\frac{1}{2} + \frac{1}{2} \frac{\gamma}{\sqrt{\gamma^2+1}}$')
ax[1].legend(frameon=False)
ax[1].set_xlabel(r'$\gamma$')
ax[1].set_ylabel(r'$\frac{\partial\gamma_\mathrm{new}}{\partial\gamma}$')

savefig('gamma_new.png')
```
